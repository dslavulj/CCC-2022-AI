# Level 3

```python
train = pd.read_csv("Data/training_data.csv", names=["sequence", "class"])
train.head()
```

```python
train_clean = train.loc[train["class"] <=1 ]
train_clean.head()
```

```python
import sklearn
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
X = train_clean["sequence"].apply(lambda x: pd.Series(list(x))).drop(columns=[30, 31, 32])
X
```

```python
test = pd.read_csv("Data/simple_test_data.csv",  names=["sequence", "class"])
test.head()
```

```python
enc = sklearn.preprocessing.OneHotEncoder()
enc.fit(X)
X = enc.transform(X).toarray()
X
```

```python
clf = RandomForestClassifier()
clf.fit(X, train_clean["class"])
```

```python
X_test = test["sequence"].apply(lambda x: pd.Series(list(x)))
X_test = enc.transform(X_test).toarray()
X_test
```

```python
y_hat = clf.predict(X_test)
y_hat.tolist()
```

```python
f = open("level_3_submission.csv", "a")
for i in y_hat.tolist():
    f.write(str(i) + "\n")
f.close()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.utils import class_weight

X_new_train, X_new_test, y_new_train, y_new_test = train_test_split(X_train, train["class"], test_size=0.20, random_state=42)
```

```python

classes_weights = class_weight.compute_sample_weight(
    class_weight='balanced',
    y=train["class"]
)
len(classes_weights)
```

```python
smallest_class_count = train["class"].sum()
largest_class_count = len(train["class"]) - smallest_class_count
spw = largest_class_count / smallest_class_count
print(spw)
```

```python
from xgboost import XGBClassifier

classes_weights = class_weight.compute_sample_weight(
    class_weight='balanced',
    y=y_new_train
)

clf = XGBClassifier(booster='gbtree', 
   			objective='multi:softmax',
			num_class=5, 
   			max_depth=30, learning_rate=0.1, 
  			n_estimators=105, 
  			scale_pos_weight=spw, 
  			random_state=101, 
            n_jobs=-1, 
			sample_weight=classes_weights,
			eval_metric="aucpr")

clf.fit(X_new_train, y_new_train, sample_weight=classes_weights)

# make predictions
preds = clf.predict(X_new_test)
print(classification_report(y_new_test, preds))
```
